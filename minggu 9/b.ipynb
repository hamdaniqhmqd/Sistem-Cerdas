```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

suhu = ctrl.Antecedent(np.arange(0, 38, 1), 'suhu')
kelembapan = ctrl.Antecedent(np.arange(0, 71, 1), 'kelembapan')
durasi = ctrl.Consequent(np.arange(0, 91, 1), 'durasi')

suhu['dingin'] = fuzz.trimf(suhu.universe, [0, 7.5, 15])
suhu['normal'] = fuzz.trimf(suhu.universe, [11, 18.5, 26])
suhu['panas'] = fuzz.trimf(suhu.universe, [22, 29.5, 37])

kelembapan['kering'] = fuzz.trimf(kelembapan.universe, [0, 10, 20])
kelembapan['cukup'] = fuzz.trimf(kelembapan.universe, [15, 32.5, 50])
kelembapan['basah'] = fuzz.trimf(kelembapan.universe, [40, 55, 70])

durasi['pendek'] = fuzz.trimf(durasi.universe, [0, 14, 28])
durasi['sedang'] = fuzz.trimf(durasi.universe, [20, 34, 48])
durasi['lama'] = fuzz.trimf(durasi.universe, [40, 65, 90])

rule1 = ctrl.Rule(suhu['panas'] & kelembapan['kering'], durasi['lama'])
rule2 = ctrl.Rule(suhu['panas'] & kelembapan['cukup'], durasi['sedang'])
rule3 = ctrl.Rule(suhu['panas'] & kelembapan['basah'], durasi['pendek'])
rule4 = ctrl.Rule(suhu['normal'] & kelembapan['kering'], durasi['lama'])
rule5 = ctrl.Rule(suhu['normal'] & kelembapan['cukup'], durasi['sedang'])
rule6 = ctrl.Rule(suhu['normal'] & kelembapan['basah'], durasi['lama'])
rule7 = ctrl.Rule(suhu['dingin'] & kelembapan['kering'], durasi['lama'])
rule8 = ctrl.Rule(suhu['dingin'] & kelembapan['cukup'], durasi['sedang'])
rule9 = ctrl.Rule(suhu['dingin'] & kelembapan['basah'], durasi['pendek'])

watering_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
watering = ctrl.ControlSystemSimulation(watering_ctrl)

# soal nomor 24
watering.input['suhu'] = 20
watering.input['kelembapan'] = 12

watering.compute()
print(f"Durasi penyiraman: {watering.output['durasi']} detik")
# durasi.view(sim=watering)

# plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

produktivitas = ctrl.Antecedent(np.arange(0, 11, 1), 'produktivitas')
kualitas = ctrl.Antecedent(np.arange(0, 6, 1), 'kualitas')
kinerja = ctrl.Consequent(np.arange(0, 11, 1), 'kinerja')

produktivitas['rendah'] = fuzz.trimf(produktivitas.universe, [0, 0, 5])
produktivitas['sedang'] = fuzz.trimf(produktivitas.universe, [0, 5, 10])
produktivitas['tinggi'] = fuzz.trimf(produktivitas.universe, [5, 10, 10])

kualitas['rendah'] = fuzz.trimf(kualitas.universe, [0, 0, 2.5])
kualitas['sedang'] = fuzz.trimf(kualitas.universe, [0, 2.5, 5])
kualitas['tinggi'] = fuzz.trimf(kualitas.universe, [2.5, 5, 5])

kinerja['very poor'] = fuzz.trimf(kinerja.universe, [0, 0, 2])
kinerja['poor'] = fuzz.trimf(kinerja.universe, [0, 2, 4])
kinerja['fair'] = fuzz.trimf(kinerja.universe, [2, 5, 8])
kinerja['good'] = fuzz.trimf(kinerja.universe, [6, 8, 10])
kinerja['very good'] = fuzz.trimf(kinerja.universe, [8, 10, 10])

rule1 = ctrl.Rule(produktivitas['rendah'] & kualitas['rendah'], kinerja['very poor'])
rule2 = ctrl.Rule(produktivitas['rendah'] & kualitas['sedang'], kinerja['poor'])
rule3 = ctrl.Rule(produktivitas['rendah'] & kualitas['tinggi'], kinerja['fair'])
rule4 = ctrl.Rule(produktivitas['sedang'] & kualitas['rendah'], kinerja['poor'])
rule5 = ctrl.Rule(produktivitas['sedang'] & kualitas['sedang'], kinerja['fair'])
rule6 = ctrl.Rule(produktivitas['sedang'] & kualitas['tinggi'], kinerja['good'])
rule7 = ctrl.Rule(produktivitas['tinggi'] & kualitas['rendah'], kinerja['fair'])
rule8 = ctrl.Rule(produktivitas['tinggi'] & kualitas['sedang'], kinerja['good'])
rule9 = ctrl.Rule(produktivitas['tinggi'] & kualitas['tinggi'], kinerja['very good'])

kinerja_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
kinerja_karyawan = ctrl.ControlSystemSimulation(kinerja_ctrl)

produktivitas_input = int(input("Masukkan skala produktivitas (0-10): "))
kualitas_input = int(input("Masukkan skala kualitas pekerjaan (0-5): "))

kinerja_karyawan.input['produktivitas'] = produktivitas_input
kinerja_karyawan.input['kualitas'] = kualitas_input

kinerja_karyawan.compute()

print(f"Kualitas kinerja karyawan : {kinerja_karyawan.output['kinerja']}")
kinerja.view(sim=kinerja_karyawan)
```
